```python
import matplotlib.pyplot as plt

# Data for bar plot
groups = ['Instrumental (Static)', 'Instrumental (Interactive)', 'Choral (Static)', 'Choral (Interactive)']
values = [62, 63, 46, 45]  # Number of students in each group

# Create the bar plot
plt.figure(figsize=(10, 6))
plt.bar(groups, values, color=['cyan', 'yellow', 'green', 'red'])

# Add titles and labels
plt.title("Student Distribution Across Groups", fontsize=16)
plt.ylabel("Number of Students", fontsize=12)
plt.xlabel("Groups", fontsize=12)

# Add gridlines for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for comparative bar plot
groups = ['Static', 'Interactive']
instrumental_values = [62, 63]  # Instrumental students in Static and Interactive groups
choral_values = [46, 45]  # Choral students in Static and Interactive groups

x = np.arange(len(groups))  # X-axis positions
width = 0.35  # Width of the bars

# Create the bar plot
plt.figure(figsize=(10, 6))
plt.bar(x - width/2, instrumental_values, width, label='Instrumental', color='red')
plt.bar(x + width/2, choral_values, width, label='Choral', color='green')

# Add titles and labels
plt.title("Comparison of Ethnic Music Teaching Through Instruments", fontsize=16)
plt.ylabel("Number of Students", fontsize=12)
plt.xlabel("Groups", fontsize=12)
plt.xticks(x, groups)
plt.legend()

# Add gridlines for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for comparative bar plot
groups = ['Static', 'Interactive']
instrumental_values = [62, 63]  # Instrumental students in Static and Interactive groups
choral_values = [46, 45]  # Choral students in Static and Interactive groups

x = np.arange(len(groups))  # X-axis positions
width = 0.35  # Width of the bars

# Create the bar plot
plt.figure(figsize=(10, 6))
plt.bar(x - width/2, instrumental_values, width, label=f'Instrumental (Static: {instrumental_values[0]}, Interactive: {instrumental_values[1]})', color='red')
plt.bar(x + width/2, choral_values, width, label=f'Choral ({choral_values[0]}, Interactive: {choral_values[1]})', color='green')

# Add titles and labels
plt.title("Comparison of Ethnic Music Teaching Through Instruments", fontsize=16)
plt.ylabel("Number of Students", fontsize=12)
plt.xlabel("Groups", fontsize=12)
plt.xticks(x, groups)
plt.legend()

# Add gridlines for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Data for comparative bar plot
groups = ['Static', 'Interactive']
instrumental_values = [62, 63]  # Instrumental students in Static and Interactive groups
choral_values = [46, 45]  # Choral students in Static and Interactive groups

x = np.arange(len(groups))  # X-axis positions
width = 0.35  # Width of the bars

# Create the bar plot
plt.figure(figsize=(10, 6))
plt.bar(x - width/2, instrumental_values, width, label='Instrumental', color='green')
plt.bar(x + width/2, choral_values, width, label='Choral', color='red')

# Add titles and labels
plt.title("Comparison of Ethnic Music Teaching Through Instruments", fontsize=16)
plt.ylabel("Number of Students", fontsize=12)
plt.xlabel("Groups", fontsize=12)
plt.xticks(x, groups)

# Add a neatly formatted legend
plt.legend(title="Group Details", loc="upper left", bbox_to_anchor=(1.05, 1), borderaxespad=0.)

# Add gridlines for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()

```
